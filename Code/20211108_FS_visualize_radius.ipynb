```python
from math import pi
def circle_area(r): 
    print ("The area of the circle with radius " + str(r) + "cm is: " + str(pi * r**2))
    area = pi * r**2
    return area*10000
arena = 11431

circle3 = plt.Circle((0.8,0.5), Circle3, color='y', clip_on=False)
circle2 = plt.Circle((0.4,0.5), Circle2, color='c')
circle1 = plt.Circle((0.1,0.5),Circle1, color ='g')



fig, ax = plt.subplots(1,1,sharex=True,dpi=400,) # note we must use plt.subplots, not plt.subplot
# note we must use plt.subplots, not plt.subplot
# (or if you have an existing figure)
# fig = plt.gcf()
# ax = fig.gca()
plt.text(0.8,0.5, int(circle_area(Circle3)))
plt.text(0.4,0.5, int(circle_area(Circle2)))
plt.text(0.075,0.5, int(circle_area(Circle1)))

cir3=circle_area(Circle3)/arena
cir2= circle_area(Circle2)/arena
cir1 = circle_area(Circle1)/arena

plt.text(0.8,0.45, round(cir3*100,1))
plt.text(0.4,0.45, round(cir2*100,1))
plt.text(0.075,0.45, round(cir1*100,1))

plt.text(0.87,0.45, "%")
plt.text(0.45,0.45,  "%")
plt.text(0.13,0.45,  "%")

plt.text(0.84,0.72, "r=20cm")
plt.text(0.42,0.67,  "r=15cm")
plt.text(0.12,0.6,  "r=7.5cm")

plt.text(0.3,0.8, "Arena area: 11431cm2")
ax.add_patch(circle3)
ax.add_patch(circle2)
ax.add_patch(circle1)
ax.axis("equal")


fig.savefig(figures+'circles.png')
```
