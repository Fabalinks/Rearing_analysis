## Create an appropriate pandas series for graphing  named appropriately ignoring nan and as float round the time to 2 decimals  (series was an object)

```python
resolution = 1
index = '[3::resolution]'
time = Arena_movement['Unnamed: 1'][3:].astype(float).round(decimals=2).rename('Time(s)')

X = Arena_movement['Arena.4'][3:].astype(float)
Y= Arena_movement ['Arena.5'][3:].astype(float)
Z= Arena_movement ['Arena.6'][3:].astype(float).rename('Arena position')



derivative = Arena_movement ['Arena.6'][3:].astype(float).rename('Arena position').diff()
Frame = Arena_movement ['Unnamed: 0'][3:].astype(float)


df = pd.merge(time,Z, right_index = True,left_index = True)


# setting first name as index column
#df.set_index(["Unnamed: 1"], inplace = True,append = True, drop = True)
   
# resetting index
#df.reset_index(inplace = True)
df

df.head(10)
```
